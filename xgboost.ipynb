```python
import numpy as np
import pandas as pd
import xgboost
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold
```

```python
stack_1 = pd.read_pickle('stack/stack_1.pkl')[0].values
stack_2 = pd.read_pickle('stack/stack_2.pkl')[0].values
stack_3 = pd.read_pickle('stack/stack_3.pkl')[0].values
```

```python
stack = []
for i in range(stack_1.shape[0]):
    s1, s2, s3 = stack_1[i], stack_2[i], stack_3[i]
    if s1 != -1: stack.append(s1)
    if s2 != -1: stack.append(s2)
    if s3 != -1: stack.append(s3)
```

```python
stack_arr = np.array(stack)
```

```python
result_1 = pd.read_pickle('stack/result_1.pkl')['category'].values
result_2 = pd.read_pickle('stack/result_2.pkl')['category'].values
result_3 = pd.read_pickle('stack/result_3.pkl')['category'].values
```

```python
result = []
for i in range(result_1.shape[0]):
    sum = result_1[i] + result_2[i] + result_3[i]
    if sum > 1: result.append(1)
    else: result.append(0)
result_arr = np.array(result)
```

```python
data_set = np.loadtxt("dataset/training_set.txt").astype(np.int64)
data = data_set[:, :2]
y = data_set[:, -1]
```

```python
classic = pd.read_pickle('features.pkl')
```

```python
classic = classic[['year1', 'year2',
                    # 'year_differ',
                     #'sameAutCount',
                     #'sameTitleWord',
                     #'sameT1Ab2',
                     #'sameT2Ab1',
                    # 'sameAbstract',
                     'T1T2_c',
                     'T1Ab2_c',
                     'T2Ab1_c',
                     'Ab1Ab2_c',
                     'aut1_sum', 'aut2_sum',#'aut_sum',
                     'connection1', 'connection2',
                     #'connection_max',
                     'len1', 'len2',
                    # 'j1', 'j2',
                    # 'year_sum',
                    # 'wmDistance'
                   'edge2',
                   'edge3',
                   'edge2_neg',
                   'edge3_neg',
                   'edge3_neg_middle',
                   'edge3_neg_sid',
                 #  'edge4'
                  # 'path',
                  # 'double_connex1',
                  # 'double_connex2',
                  # 'nocom_connex1',
                  # 'nocom_connex2',
                   'edge2_weighted',
                  # 'sameAut',
                  # 'AAA_edge',
                  #  'ADA_edge',
                  # '=A=D=A=',
                   'edge3_weighted',
                   #'sameJ',
                   #'edge2_123_weighted',
                   #'edge2_sqr_weighted',
                   #'connex_max','connex_min',
                   #'year_max','year_min',
                   '->D->',#'->D<-','<-D->',
                   #'->D->_w',
                   #'->D->D->',
                   #'src1','tgt2',
                    ]]
```

```python
classic['edge23'] = list(map(lambda x: 50*x[0] + x[1], zip(classic['edge2'].values.tolist(), classic['edge3'].values.tolist())))
```

```python
classic['rnn'] = stack_arr
```

```python
lsa1 = pd.read_pickle('lsa1.pkl')
lsa2 = pd.read_pickle('lsa2.pkl')
#X = np.concatenate((classic.values, lsa1.values[:,:15], lsa2.values[:,:15]), axis=1)
```

```python
d2v1 = pd.read_pickle('d2v1.pkl')
d2v2 = pd.read_pickle('d2v2.pkl')
X = np.concatenate((classic.values, lsa1.values[:,:15], lsa2.values[:,:15], d2v1.values[:,:0], d2v2.values[:,:0]), axis=1)
```

```python
seed = 7
test_size = 0.15
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
```

```python
X_train.shape
```

```python
model = XGBClassifier(max_depth=12, n_jobs = 4, objective='binary:logistic')
model.fit(X_train, y_train)
```

```python
model.save_model('xgb.model')
```

```python
fscore = model.feature_importances_
```

```python
fscore
```

```python
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.3f%%" % (accuracy * 100.0))
```
