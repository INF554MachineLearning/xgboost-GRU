```python
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import stopwords
```

** Data import **

```python
columes = ["id", "year", "title", "authors", "journal", "abstract"]
df= pd.read_csv("dataset/node_information.csv",names=columes)

data_dic = {}
for index, row in df.iterrows():
    data_dic[row['id']] = (row['year'], row['title'], row['authors'], row['journal'], row['abstract'])
    
stop = set(stopwords.words('english')+['1','2','3','4','5','6','7','8','9','0'])
```

** cleaning the title and abstract **

```python
corpus_title =df["title"].values.tolist()
corpus_abstract = df["abstract"].values.tolist()
tok_corp_tit = [nltk.word_tokenize(sent) for sent in corpus_title]
tok_corp_abs = [nltk.word_tokenize(sent) for sent in corpus_abstract]
```

```python
corpus_title_c = []
for sent in tok_corp_tit:
    temp = ''
    for w in sent:
        if w not in stop:
            temp = temp + ' ' + w
    corpus_title_c.append(temp)
clean_title = pd.DataFrame()
clean_title['text'] = corpus_title_c
clean_title.to_pickle('clean_title.pkl')
```

```python
corpus_abstract_c = []
for sent in tok_corp_abs:
    temp = ''
    for w in sent:
        if w not in stop:
            temp = temp + ' ' + w
    corpus_abstract_c.append(temp)
clean_abstract = pd.DataFrame()
clean_abstract['text'] = corpus_abstract_c
clean_abstract.to_pickle('clean_abstract.pkl')
```
